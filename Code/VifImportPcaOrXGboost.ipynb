```python
# Specify the path to the CSV file in the parent directory
file_path = "../Dataset/pd_speech_features/pd_speech_features.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path, sep=',',skiprows=1)
df=df.drop('id',axis=1)
```

# SMOTE

```python
X=df.drop('class',axis=1)
y=df['class']
smote_df = SMOTE(sampling_strategy='auto', random_state=42)
X_smote, y_smote = smote_df.fit_resample(X, y)
```

```python
unique, counts = np.unique(y, return_counts=True)
original_class_distribution = dict(zip(unique, counts))
unique, counts = np.unique(y_smote, return_counts=True)
resampled_class_distribution = dict(zip(unique, counts))

print("Original class distribution:", original_class_distribution)
print("Resampled class distribution:", resampled_class_distribution)
```

# Normalization

```python
features = X_smote.columns

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Scale the features
X_smote_scaled = pd.DataFrame(scaler.fit_transform(X_smote[features]), columns=features)
```

# Import Vif Dataset

```python
# Read the CSV file into a DataFrame
X_smote_scaled_vif = pd.read_csv("X_smote_scaled_vif.csv", sep=',')
X_smote_scaled_vif.shape, y_smote.shape
```

# PCA

```python
# Apply PCA
pca = PCA(n_components=0.95)
X_smote_scaled_vif_pca = pca.fit_transform(X_smote_scaled_vif)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Find the number of components explaining 95% variance
n_components = np.argmax(cumulative_variance >= 0.95) + 1

# Plot cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(cumulative_variance, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Number of Components')
plt.axvline(x=n_components, color='red', linestyle='--', label=f'{n_components} components (95% variance)')
plt.legend()
plt.grid(True)
plt.show()
```

# XGBoost Feature selection

```python
X_train, X_test, y_train, y_test=train_test_split(X_smote_scaled_vif,y_smote,test_size=0.2, random_state=42)

xgb_classifier = xgb.XGBClassifier(n_estimators=100, random_state=42)

xgb_classifier.fit(X_train, y_train)

threshold_values = [0.001,
                    0.0015,
                    0.002,
                    0.0025,
                    0.003,
                    0.0035,
                    0.004,
                    0.0045,
                    0.005,
                    0.0055,
                    0.006,
                    0.0065,
                    0.007,
                    0.0075,
                    0.008,
                    0.0085,
                    0.009,
                    0.0095,
                    0.001,
                    0.0015,
                    0.002]

accuracy_dict_xgb = {}
threshold_better_xgb = 0
accuracy_better_xgb = 0
for threshold in threshold_values:
    # Create a feature selector based on the current threshold
    feature_selector = SelectFromModel(xgb_classifier, threshold=threshold, prefit=True)
    
    # Transform the training data using the selector
    X_train_selected = feature_selector.transform(X_train)
    
    # Train an XGBoost model with the selected features
    xgb_selected = xgb.XGBClassifier(n_estimators=100, random_state=42)
    xgb_selected.fit(X_train_selected, y_train)
    
    # Transform the testing data using the same selector
    X_test_selected = feature_selector.transform(X_test)
    
    # Make predictions with the selected features
    y_pred = xgb_selected.predict(X_test_selected)
    
    # Calculate the accuracy and store it in the dictionary
    accuracy_xgb = accuracy_score(y_test, y_pred)
    accuracy_dict_xgb[threshold] = accuracy_xgb
    
    if accuracy_xgb > accuracy_better_xgb:
        threshold_better = threshold
        selected_feature_mask_xgb = feature_selector.get_support()
        selected_feature_names_xgb = X_train.columns[selected_feature_mask_xgb]
        accuracy_better_xgb = accuracy_xgb
   
        print(threshold)
        print(len(selected_feature_names_xgb))
        print(selected_feature_names_xgb)
        print(accuracy_xgb)
        print("")
    
    

    
```

# PCA

```python
# Create base models
knn_model = KNeighborsClassifier()
gb_model = GradientBoostingClassifier(random_state=42)
svm_model = SVC(probability=True, random_state=42)
xgb_model = XGBClassifier(random_state=42)

# Set up parameter grids for hyperparameter tuning
knn_param_grid = {'n_neighbors': [3, 5, 7],
                'weights': ['uniform', 'distance'],
                'p': [1, 2]}
gb_param_grid = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]}
svm_param_grid = {'C': [0.1, 1, 10],
                'kernel': ['linear', 'rbf', 'poly'],
                'gamma': ['scale', 'auto']}
xgb_param_grid = {
                'learning_rate': [0.1, 0.01],
                'max_depth': [3, 5, 7],
                'n_estimators': [100, 200, 300]}

start = time.time()

# Perform grid search for each base model
knn_grid = GridSearchCV(knn_model, knn_param_grid, cv=5)
gb_grid = GridSearchCV(gb_model, gb_param_grid, cv=5)
svm_grid = GridSearchCV(svm_model, svm_param_grid, cv=5)
xgb_grid = GridSearchCV(xgb_model, knn_param_grid, cv=5)

# Fit the base models with the best hyperparameters on the training data
knn_grid.fit(X_train, y_train)
gb_grid.fit(X_train, y_train)
svm_grid.fit(X_train, y_train)
xgb_grid.fit(X_train, y_train)

# Obtain cross-validated predictions for each base model on the training set
knn_pred_train = cross_val_predict(knn_grid.best_estimator_, X_train, y_train, cv=5, method='predict_proba')[:, 1]
gb_pred_train = cross_val_predict(gb_grid.best_estimator_, X_train, y_train, cv=5, method='predict_proba')[:, 1]
svm_pred_train = cross_val_predict(svm_grid.best_estimator_, X_train, y_train, cv=5, method='predict_proba')[:, 1]
xgb_pred_train = cross_val_predict(xgb_grid.best_estimator_, X_train, y_train, cv=5, method='predict_proba')[:, 1]

# Create a new training set with predictions from the base models
stacked_X_train = np.column_stack((knn_pred_train, gb_pred_train, svm_pred_train, xgb_pred_train))

# Create a meta-model (e.g., XGBoost) to make the final prediction
meta_model = LogisticRegression()
meta_model.fit(stacked_X_train, y_train)

# Use the base models to make predictions on the test set
knn_pred_test = knn_grid.best_estimator_.predict_proba(X_test)[:, 1]
gb_pred_test = gb_grid.best_estimator_.predict_proba(X_test)[:, 1]
svm_pred_test = svm_grid.best_estimator_.predict_proba(X_test)[:, 1]
xgb_pred_test = xgb_grid.best_estimator_.predict_proba(X_test)[:, 1]

# Create a new test set for stacking using predictions from base models
stacked_X_test = np.column_stack((knn_pred_test, gb_pred_test, svm_pred_test, xgb_pred_test))

# Use the meta-model to make the final prediction on the stacked test data
final_prediction_test = meta_model.predict(stacked_X_test)

# Evaluate the performance of the stacked model on the test set
performance_metrics = {}
accuracy = accuracy_score(y_test, final_prediction_test)
precision = precision_score(y_test, final_prediction_test)
recall = recall_score(y_test, final_prediction_test)
f1 = f1_score(y_test, final_prediction_test)
auc = roc_auc_score(y_test, final_prediction_test)
time_execution = time.time() - start
performance_metrics["ensemble"] = {
    'accuracy': accuracy,
    'precision': precision,
    'recall': recall,
    'f1': f1,
    'auc':auc,
    'time_execution':time_execution
}
print(performance_metrics)
```
